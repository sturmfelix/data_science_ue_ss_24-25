## Daten in `pandas` laden

```python
from pandas import read_csv
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = read_csv(
    "../data/fake_profiles.csv", 
    sep=",", 
    decimal=".", 
    usecols=range(1, 13)
    )
```

## Einfache statistische Auswertungen

```python
data.info()
```

```python
data.describe()
```

```python
data["words_fullname"].mean()
```

## Visualisierung Pandas

```python
data.plot()
```

```python
data["words_fullname"].plot(kind="hist")
```

```python
data["words_fullname"].plot(kind="box")
```

```python
data["profile_pic"].value_counts().plot(
    kind="bar",
    title="Profile has Profile Picture",
    xlabel="Available Categories",
    ylabel="Number of Profiles",
    color=["black", "grey"]
)
```

```python
data.plot(
    kind="scatter",
    x="num_posts",
    y="num_followers"
)
```

## Praktische Plots die man kennen sollte

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue="is_fake")
```

```python
sns.heatmap(
    data.select_dtypes(include="number").corr()
)
```

```python
sns.clustermap(
    data.select_dtypes(include="number").corr()
)
```
