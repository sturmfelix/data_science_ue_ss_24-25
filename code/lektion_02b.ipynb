````python
from pandas import read_csv
````

````python
data = read_csv("../data/fake_profiles.csv", index_col=0)
````

````python
data.head(3)
````

````python
data.info()
````

````python
features = ["rel_num_numeric_char_username", 
            "words_fullname", 
            "rel_num_numeric_char_fullname", 
            "description_length"
]

targets = ["is_fake"]

X = data[features]
y = data[targets]
````

````python
from sklearn.cluster import KMeans, AgglomerativeClustering
````

## K-Means

````python
kmeans_engine = KMeans(n_clusters=2)
````

````python
kmeans_engine.n_clusters
````

````python
kmeans_engine.fit(X)
````

````python
kmeans_engine.predict(X)
````

````python
kmeans_engine.fit_predict(X)
````

## Agglomerative (Hierarchical) Clustering

````python
hierarchical_engine = AgglomerativeClustering(n_clusters=3)
````

````python
clusters = hierarchical_engine.fit_predict(X)
````

````python
df = data.copy()
df["cluster"] = clusters
````

````python
df
````

````python
df.plot(
    kind = "scatter", 
    x="description_length", 
    y="words_fullname", 
    c="cluster", 
    cmap="inferno"
)
````

## Scipy Clustering

````python
from scipy.cluster import hierarchy
from matplotlib import pyplot as plt
````

````python
distances = hierarchy.linkage(X, method="ward")
````

````python
hierarchy.dendrogram(distances)
plt.show()
````
